```python
import csv
import re
import sqlite3


list_tuples = []
list_lists = []
header_tuple = None
header_list = None
column_types = []

with open('LMPD_OP_BIAS_7.csv') as rawhcdata:
    reader = csv.reader(rawhcdata)
    for row in reader:
        stripped_row = []
        for entry in row:
            stripped_row.append(entry.strip().upper())
        list_tuples.append(tuple(stripped_row))
        list_lists.append(stripped_row)
    header_tuple = list_tuples.pop(0)
    header_list = list_lists.pop(0)

print(header_list)
print(list_lists)
for value in header_list:
    print(value)
print(len(header_list))
print(len(list_lists))

counter = 0
while counter < len(header_list):
    int_row_counter = 1
    for row in list_lists:
        if bool(re.search(r'^\d+$', row[counter])) and int_row_counter < len(list_lists):
            int_row_counter += 1
            continue
        elif bool(re.search(r'^\d+$', row[counter])) and int_row_counter >= len(list_lists):
            column_types.append('INTEGER PRIMARY KEY')
            break
        else:
            column_types.append('TEXT')
            break
    counter += 1
print(column_types)

# incident_numbers = []
# for row in list_lists:
#     incident_numbers.append(row[1])
# print(len(incident_numbers))
# print(len(set(incident_numbers)))
# if len(incident_numbers) == len(set(incident_numbers)):
#     column_types[1] += " UNIQUE"

new_counter = 0
statement_center = ''
while new_counter < len(header_list):
    statement_center += ("\n    "+ header_list[new_counter] + " " + column_types[new_counter] + ",")
    new_counter += 1
statement_center = statement_center.rstrip(',')
create_statement = "CREATE TABLE IF NOT EXISTS LOU_HATE_CRIMES(" + statement_center + "\n);"
print(create_statement)



conn = sqlite3.connect('lou_hate_crime_database.db')
cur = conn.cursor()
try:
    cur.execute("DROP TABLE IF EXISTS LOU_HATE_CRIMES;")
    cur.execute(create_statement)
    cur.executemany("INSERT INTO LOU_HATE_CRIMES VALUES(?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", list_tuples)
    conn.commit()
finally:
    conn.rollback()
    cur.close()
    conn.close()
```
